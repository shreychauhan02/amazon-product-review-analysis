```python
import requests
import json
import pandas as pd
```

```python
url = "https://huggingface.co/datasets/McAuley-Lab/Amazon-Reviews-2023/resolve/main/raw/review_categories/Office_Products.jsonl"

rows = []
count = 0
MAX_ROWS = 53258   

with requests.get(url, stream=True) as r:
    for line in r.iter_lines():
        if not line:
            continue
        count += 1
        record = json.loads(line)
        if record["timestamp"] >= 1420070400000:  # Jan 1, 2015
            rows.append(record)
            if len(rows) >= MAX_ROWS:
                break

df_new = pd.DataFrame(rows)
print("done. final shape:", df_new.shape)
```

```python
def black_friday_date(year):
    nov1 = pd.Timestamp(year=year, month=11, day=1)
    first_thursday = nov1 + pd.Timedelta(days=(3 - nov1.dayofweek) % 7)
    fourth_thursday = first_thursday + pd.Timedelta(weeks=3)
    return fourth_thursday + pd.Timedelta(days=1)

def get_sale_windows(year):
    bf = black_friday_date(year)
    return {
        "black_friday": (bf, bf),
        "cyber_monday": (bf + pd.Timedelta(days=3), bf + pd.Timedelta(days=3)),
        "thanksgiving_week": (bf - pd.Timedelta(days=3), bf + pd.Timedelta(days=4)),
        "holiday_season": (pd.Timestamp(year, 12, 1), pd.Timestamp(year, 12, 31)),
        "prime_day": (pd.Timestamp(year, 7, 11), pd.Timestamp(year, 7, 13)) if year >= 2015 else None,
    }

for col in ["is_black_friday", "is_cyber_monday", "is_thanksgiving_week", "is_holiday_season", "is_prime_day"]:
    df_new[col] = False

for year in df_new["review_year"].unique():
    windows = get_sale_windows(year)
    mask_year = df_new["review_year"] == year

    for col, key in [
        ("is_black_friday", "black_friday"),
        ("is_cyber_monday", "cyber_monday"),
        ("is_thanksgiving_week", "thanksgiving_week"),
        ("is_holiday_season", "holiday_season"),
        ("is_prime_day", "prime_day"),
    ]:
        window = windows[key]
        if window is None:
            continue
        start, end = window
        in_window = df_new["review_date"].between(start, end)
        df_new.loc[mask_year & in_window, col] = True

df_new["is_sale_period"] = df_new[
    ["is_black_friday", "is_cyber_monday", "is_thanksgiving_week", "is_holiday_season", "is_prime_day"]
].any(axis=1)

print(df_new[["is_black_friday","is_cyber_monday","is_thanksgiving_week","is_holiday_season","is_prime_day","is_sale_period"]].sum())
```
